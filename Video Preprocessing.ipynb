### Now the videos are downloaded and in total there are 11,633 videos for different signs so now we will be preprocessing it by capturing frames from the videos and then converting it into images and in addition assigning them their classes.

```python
import cv2     # for capturing videos
import math   # for mathematical operations
import matplotlib.pyplot as plt    # for plotting the images
%matplotlib inline
import pandas as pd
from keras.preprocessing import image   # for preprocessing the images
import numpy as np    # for mathematical operations
from keras.utils import np_utils
from skimage.transform import resize   # for resizing images
from sklearn.model_selection import train_test_split
from glob import glob
from tqdm import tqdm
import os
import re
from imutils import paths
```

```python
vid=[]
for file in os.listdir('D:\GAIP\Project\Data'):
    vid.append(os.getcwd()+"/"+"Data/"+file)
train=pd.DataFrame(data=vid,columns=['video'])
```

```python
train.head(3)
```

```python
train['video'] = train['video'].replace('\\', '/')
```

```python
import time

cap= cv2.VideoCapture(train.video[0])


fps= int(cap.get(cv2.CAP_PROP_FPS))

print("This is the fps ", fps)



while cap.isOpened():
    ret,frame= cap.read()

    if ret == True:

        gray= cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        time.sleep(1/fps)

        cv2.imshow('Video in Grayscale', gray)

        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    else:
        break


cap.release()
cv2.destroyAllWindows()
```

```python
for vid in train['video']:
    print(vid.split('/')[2].split('.')[0])
```

```python
# storing the frames from training videos
for i in tqdm(range(0,len(train))):
    count = 0
    videoFile = train['video'][i]
    cap = cv2.VideoCapture(videoFile)   # capturing the video from the given path
    frameRate = cap.get(5) #frame rate
    if (cap.isOpened()==True):
        while(cap.isOpened()):
            frameId = cap.get(1) #current frame number
            ret, frame = cap.read()
            if (ret != True):
                break
            if (frameId % math.floor(frameRate) == 0):
            # storing the frames in a new folder named train_1
                filename ='./train_1/'+videoFile.split('/')[2].split('.')[0]+"_frame%d.jpg" % count
                cv2.imwrite(filename, frame)
                count+=1
        cap.release()
```

```python
images = glob("train_1/*.jpg")
train_image = []
train_class = []
for i in tqdm(range(len(images))):
    s=images[i].split('_')[1].split('\\')[1]
    print(re.split('\d+',s)[0])
```

```python
# getting the names of all the images
images = glob("train_1/*.jpg")
train_image = []
train_class = []
for i in tqdm(range(len(images))):
    # creating the image name
    train_image.append(images[i].split('\\')[1])
    # creating the class of image
    s=images[i].split('_')[1].split('\\')[1]
    train_class.append(re.split('\d+',s)[0])
    
# storing the images and their class in a dataframe
train_data = pd.DataFrame()
train_data['image'] = train_image
train_data['class'] = train_class

# converting the dataframe into csv file 
train_data.to_csv('train_new.csv',header=True, index=False)
```
